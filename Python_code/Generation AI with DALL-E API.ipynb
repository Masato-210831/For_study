```python
from openai import OpenAI

client = OpenAI(
    # githubに上げるためシークレット
    api_key="secret"
)
```

```python
def generate_image(prompt):
    response = client.images.generate(
        
        model = 'dall-e-2', #エンジンを指定
        prompt=prompt,
        n=1, #生成するレスポンス数
        size='256x256', #画像サイズ
        quality='standard'
    )
    
    return response.data[0].url
```

# GMMによるデータ生成

```python
mus = np.array([[2.0, 54.50],
                [4.3, 80.0]])
covs = np.array([[[0.07, 0.44],
                  [0.44, 33.7]],
                 [[0.17, 0.94],
                  [0.94, 36.00 ]]])
phis = np.array([0.35, 0.65])
```

```python
def sample():
    k = np.random.choice(2, p=phis) # 35%で0, 65%で1を返す
    mu, cov = mus[k], covs[k]
    x = np.random.multivariate_normal(mu, cov)
    return x

sample()
```

```python
N = 500
xs = np.zeros((N, 2))
for i in range(N):
     xs[i] = sample()

plt.scatter(xs[:,0], xs[:,1], color='darkred', alpha=0.7)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

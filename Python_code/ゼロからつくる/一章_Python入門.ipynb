## Matplotlib

```python
x = np.arange(0, 6, 0.1)
y_1 = np.sin(x)
```

```python
plt.plot(x, y_1)
```

```python
# cos
y_2 = np.cos(x)

plt.plot(x, y_1, label='sin')
plt.plot(x, y_2, linestyle="--", label='cos')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin & cos')
plt.legend(loc='upper right')
```

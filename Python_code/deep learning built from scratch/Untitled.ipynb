```python
import numpy as np
from common.functions import softmax, cross_entropy_error


class simpleNet:
    def __init__(self):
        self.W = np.array([[1, 2, 3], [4, 5, 6]])

    def predict(self, x):
        return np.dot(x, self.W)

    def loss(self, x, t):
        z = self.predict(x)
        y = softmax(z)
        loss = cross_entropy_error(y, t)

        return loss

x = np.array([0.6, 0.9])
t = np.array([0, 0, 1])

net = simpleNet()

def f(W):
    print(f"x引数：{x}, t引数:{t}\n")
    print(f"\nparameter: {net.W}\n")
    print("loss:", net.loss(x ,t))
    return net.loss(x, t)
numerical_gradient(f, net.W)
```

```python
# def numerical_gradient(f, x):

def numerical_gradient(f, x):
    h = 5 # 0.0001
    grad = np.zeros_like(x)
    print("1st_x:", x)
    print("1st_grad:", grad)
    
    it = np.nditer(x, flags=['multi_index'], op_flags=['readwrite'])
    while not it.finished:
        print("重み:", x)
        idx = it.multi_index
        tmp_val = x[idx]
        x[idx] = tmp_val + h
        print("h+の重み:", x)
        fxh1 = f(x) # f(x+h)

        
        x[idx] = tmp_val - h 
        fxh2 = f(x) # f(x-h)
        grad[idx] = (fxh1 - fxh2) / (2*h)
        print(f"fxh1: {fxh1}\nfxh2:{fxh2}\ngrad:{grad[idx]}")
        print("-"*50)
        
        x[idx] = tmp_val # 値を元に戻す
        it.iternext()   
        
    return grad
```

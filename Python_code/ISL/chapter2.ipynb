## ミュータブル

```python
x = np.array([1, 2, 3, 4, 5, 6])
x_reshape = x.reshape(2, 3)
print(x)
print(x_reshape)
```

```python
print('x_reshape前のx:\n', x)
print('x_reshape:\n', x_reshape)
x_reshape[0, 0] = 5
print('左上を変更した後のx_reshape:\n',
x_reshape)
print('左上を変更した後のx:\n', x)
```

---
＜ポイント＞  
・xとx_reshapeは同じメモリに格納されているのでどちらも変更されている

# For Loop

### String Formatting

```python
# データフレームの列をループして各列の欠損値の割合を表示
# 20％が欠損値（np.nanに設定）されているデータフレームDを作成

# シード値の設定
rng = np.random.default_rng(1)
A = rng.standard_normal((127, 5))
M = rng.choice([0, np.nan], p = [0.8, 0.2], size = A.shape )

# (注意)　nan + 実数 = nan になる
A += M

# dfの作成
D = pd.DataFrame(A, columns=['fod', 'bar', 'pickle', 'snack', 'popcorn'])

D.head(3)
```

```python
# 各カラムの欠損率の表示


nan_percent_per_col = D.isna().mean() * 100

# .2%では小数点下2桁の有効数字で元の数に100を掛けて%の値で表示する
template = "columns {0} has {1:.2%} missing values"
for col in D.columns:
    print(template.format(col, nan_percent_per_col[col]))
    
```

```python
nan_percent_per_col = D.isna().mean() * 100

template = "columns {0} has {1:.2f} missing values"
for col in D.columns:
    print(template.format(col, nan_percent_per_col[col]))
```

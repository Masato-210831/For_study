# 演習問題

### 理論編
(1)  
&emsp;(a)柔軟な統計的学習法：特徴量が少なく、サンプルが多いのでvarianceを高めれば過学習をおさせてfitする  
&emsp;(b)柔軟ではない統計的学習法:pが高く、サンプルが少ないのでvalianceを高めると過学習を起こしやすい  
&emsp;(c)柔軟な統計的学習法:非線形なのでbiasの高い線形では対応できない  
&emsp;(c)柔軟でない統計的学習法:誤差項が大きいので、valianceを高めると真のfに全くfitしないため  
<br>
(2)  
&emsp;(a)回帰問題、推論、n=500, p=4  
&emsp;(b)分類問題、予想、n=20, p=13  
&emsp;(c)回帰問題, 予想, n=52, p=4  

<br>


```python
df = pd.DataFrame({'X1': [0, 2, 0, 0, -1, 1],
                   'X2': [3, 0, 1, 1, 0, -1],
                   'X3': [0, 0, 3, 2, 1, 1],
                   'Y': ['Red', 'Red', 'Red', 'Green', 'Green', 'Red']})
```

```python
# (b) K=1
min_idx = distance.argsort()[0]
min_value = distance[min_idx] 
pred_Y = df['Y'][min_idx]
print(f'x1=x2=x3=0の時の最も近いデータ\n観測値： {min_idx}\nユークリッド距離: {min_value}\n予想されるラベル：　{pred_Y}')
```

```python
# データの準備

boston_df = pd.read_csv('data/housing.csv', sep='\s+', header=None)
set_columns = ["crim","zn","indus","chas","nox","rm","age","dis","rad","tax","ptratio","b","lstat","medv"]
set_columns = list(map(lambda x: x.upper(),set_columns))
boston_df.columns = set_columns
```

```python
sns.pairplot(boston_df, y_vars='CRIM')
```

```python
boston_df['CRIM'].hist();
```

```python
pd.cut(boston_df['CRIM'], 20).value_counts()
```

```python
re_boston_df = boston_df[boston_df['CRIM'] < 50]

sns.pairplot(re_boston_df, y_vars='CRIM')
```

```python
boston_df['CRIM_flg'] = boston_df['CRIM'].map(lambda x: 'low' if x <= 4.5 else 'middle' if x < 30 else 'high')

sns.pairplot(boston_df, y_vars='CRIM', hue='CRIM_flg')
```

CRIM - 一人当たりの町の犯罪率  
ZN - 25,000平方フィート以上の区画に指定された住宅用地の割合  
INDUS - 町ごとの非小売業務用地の割合  
CHAS - チャールズ川ダミー変数（河川に沿っている場合は1、そうでない場合は0）  
NOX - 一酸化窒素濃度（1000万分の1単位）  
RM - 住宅1戸当たりの平均部屋数  
AGE - 1940年以前に建てられた自家所有住宅の割合  
DIS - ボストンの5つの雇用センターへの加重距離  
RAD - 放射状高速道路へのアクセス指数  
TAX - $10,000あたりの固定資産税率

PTRATIO - 町ごとの生徒-教師比率  
B - "1000(Bk - 0.63)^2" ここでBkは町ごとの黒人人口割合  
LSTAT - 低所得者層の人口割合  
MEDV - 自家所有住宅の中央値（$1000単位）

```python
from PIL import ImageDraw
from scipy import misc, ndimage
```

```python
raccoon = misc.face()
plt.imshow(raccoon)
```

```python
gaussian_img = ndimage.gaussian_filter(input=raccoon, sigma=20)
plt.imshow(gaussian_img)
```

```python
raccoon_img = Image.fromarray(raccoon)
img = np.array(raccoon_img.convert('L'), 'f')
sigma = 5
img_x = ndimage.gaussian_filter(input=img, sigma=(sigma, sigma), order=(0, 1))

plt.imshow(img_x, cmap='Greys_r')
```

```python
raccoon_img = Image.fromarray(raccoon)
img = np.array(raccoon_img.convert('L'), 'f')
sigma = 5
img_x = ndimage.gaussian_filter(input=img, sigma=(sigma, sigma), order=(1, 0))

plt.imshow(img_x, cmap='Greys_r')
```
